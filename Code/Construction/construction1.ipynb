```python
#@title Packages

import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import re
import warnings

warnings.simplefilter(action='ignore', category=pd.errors.PerformanceWarning)
```

```python
#@title Dataframe with S&P500 Stocks (From 5 Years Ago) and Their Basic Information

#https://www.kaggle.com/datasets/paytonfisher/sp-500-companies-with-financial-information
df_basic = pd.read_csv("../Data/sp500.csv")
df_basic = df_basic[['Symbol']]
print(df_basic["Symbol"].nunique())
```

```python
#@title Get All Google Finance Formulas

symbols = df_basic["Symbol"].unique()
attribute = "all"
end_date_str = "2025-05-30"
start_date_str = "2020-04-01"
end_date = datetime.strptime(end_date_str, "%Y-%m-%d").date()
start_date = datetime.strptime(start_date_str, "%Y-%m-%d").date()
interval = "daily"

formulas = [
    f'=GOOGLEFINANCE("{symbol}"; "{attribute}"; "{start_date}"; "{end_date}"; "{interval}")'
    for symbol in symbols
]

file_path = "../Data/formulas.txt"
with open(file_path, "w") as file:
    for formula in formulas:
        file.write(formula + "\n")  
```
