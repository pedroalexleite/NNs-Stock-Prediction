```python
#@title Packages

import pandas as pd
import numpy as np
import ast 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.colors as mcolors
import seaborn as sns
from scipy.stats import zscore
from scipy.stats import mstats
from scipy.signal import savgol_filter
from scipy.signal import periodogram
from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm
from statsmodels.tsa.stattools import kpss
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import joblib
import os
import warnings
```

```python
#@title Merge Symbol Characteristics with Performance

characteristics_df = pd.read_csv('/Users/pedroalexleite/Desktop/Tese/Dados/stocks_characteristics.csv')
results_df = pd.read_csv('/Users/pedroalexleite/Desktop/Tese/Dados/final_results.csv')
merged_df = pd.merge(results_df[['Symbol', 'RMSE', 'Accuracy']], characteristics_df, on='Symbol', how='left')

merged_df_norm = merged_df.copy()
numeric_cols = merged_df.select_dtypes(include=['number']).columns
scaler = MinMaxScaler(feature_range=(0, 1))
merged_df_norm[numeric_cols] = scaler.fit_transform(merged_df[numeric_cols])

merged_df_norm.head()
```

```python
#@title Analyze Performance for each Symbol

variables = ['RMSE', 'Accuracy']

for column in variables:
    print(f"\n{'='*50}")
    print(f"{column.upper()}")
    print(f"{'='*50}")
    
    print(f"\nTop 5 HIGHEST {column}:")
    print("-" * 30)
    top_max = merged_df.nlargest(5, column)[['Symbol', column]]
    for idx, row in top_max.iterrows():
        print(f"{row['Symbol']:>8}: {row[column]:>12.4f}")
    
    print(f"\nTop 5 LOWEST {column}:")
    print("-" * 30)
    top_min = merged_df.nsmallest(5, column)[['Symbol', column]]
    for idx, row in top_min.iterrows():
        print(f"{row['Symbol']:>8}: {row[column]:>12.4f}")
```

```python
#@title Correlation for Characteristics and Performance Each Symbol

custom_cmap = mcolors.LinearSegmentedColormap.from_list("custom_cmap", ['#4285f4', '#ea4335'])
corr_matrix = merged_df_norm.drop(columns=['Symbol']).corr(numeric_only=True)
rmse_accuracy_corr = corr_matrix.loc[['RMSE', 'Accuracy']]

plt.figure(figsize=(12, 6))
sns.heatmap(
    rmse_accuracy_corr,
    annot=False,
    cmap=custom_cmap,
    fmt='.2f',
    linewidths=0.5,
    cbar_kws={"orientation": "vertical"}
)
plt.xticks(rotation=90, fontsize=8)
plt.yticks(rotation=0, fontsize=8)
plt.gca().tick_params(axis='x', which='both', bottom=True, top=False)
plt.gca().spines['top'].set_visible(False)
labels = [item.get_text() for item in plt.gca().get_xticklabels()]
plt.gca().set_xticklabels(labels, fontsize=6)
plt.show()
```

```python
#@title Heatmap for Characteristics and Performance Each Symbol

custom_cmap = mcolors.LinearSegmentedColormap.from_list("custom_cmap", ['#4285f4', '#ea4335'])
rmse_accuracy_data = merged_df_norm[['Symbol', 'RMSE', 'Accuracy']]

plt.figure(figsize=(12, 8))
sns.heatmap(
    rmse_accuracy_data.drop(columns=['Symbol']).set_index(merged_df['Symbol']),
    cmap=custom_cmap,
    cbar=True,
    linewidths=0.5,
    linecolor='gray',
    annot=False
)
plt.xticks(rotation=90, ha='right', fontsize=8)
plt.yticks(rotation=0, fontsize=8)
plt.show()
```

```python
#@title Extract Characteristics and Performance for each Sector

sp500 = pd.read_csv("/Users/pedroalexleite/Desktop/Tese/Dados/sp500.csv")
merged_df_2 = pd.merge(merged_df, sp500[['Symbol', 'Sector']], on='Symbol', how='left')

group_cols = ['RMSE', 'Accuracy', 'Price', 'Volume', 'Growth', 'Volatility', 'Seasonality', 'Sharpe Ratio', 'Max Drawdown']
sectors_df = merged_df_2.groupby('Sector')[group_cols].mean().reset_index()

sectors_df_norm = sectors_df.copy()
numeric_cols_2 = sectors_df.select_dtypes(include=['number']).columns
scaler_2 = MinMaxScaler(feature_range=(0, 1))
sectors_df_norm[numeric_cols_2] = scaler_2.fit_transform(sectors_df[numeric_cols])

sectors_df_norm.head()
```

```python
#@title Correlation for Characteristics and Performance Each Sector

custom_cmap = mcolors.LinearSegmentedColormap.from_list("custom_cmap", ['#4285f4', '#ea4335'])
corr_matrix = sectors_df_norm.drop(columns=['Sector']).corr(numeric_only=True)
rmse_accuracy_corr = corr_matrix.loc[['RMSE', 'Accuracy']]

plt.figure(figsize=(12, 6))
sns.heatmap(
    rmse_accuracy_corr,
    annot=False,
    cmap=custom_cmap,
    fmt='.2f',
    linewidths=0.5,
    cbar_kws={"orientation": "vertical"}
)

plt.xticks(rotation=90, fontsize=8)
plt.yticks(rotation=0, fontsize=8)
plt.gca().tick_params(axis='x', which='both', bottom=True, top=False)
plt.gca().spines['top'].set_visible(False)
labels = [item.get_text() for item in plt.gca().get_xticklabels()]
plt.gca().set_xticklabels(labels, fontsize=6)
plt.show()
```

```python
#@title Heatmap for Characteristics and Performance of Each Sector

custom_cmap = mcolors.LinearSegmentedColormap.from_list("custom_cmap", ['#4285f4', '#ea4335'])
rmse_accuracy_data = sectors_df_norm[['Sector', 'RMSE', 'Accuracy']]

plt.figure(figsize=(12, 8))
sns.heatmap(
    rmse_accuracy_data.drop(columns=['Sector']).set_index(sectors_df_norm['Sector']),
    cmap=custom_cmap,
    cbar=True,
    linewidths=0.5,
    linecolor='gray',
    annot=False
)
plt.xticks(rotation=90, fontsize=8)
plt.yticks(rotation=0, fontsize=8)
plt.show()
```

```python
#@title Plot Risk-Return

risk_return_df = pd.read_csv("/Users/pedroalexleite/Desktop/Tese/Dados/risk_return.csv")
unique_sectors = risk_return_df['Sector'].unique()

sector_colors = {
    'Consumer Discretionary': '#4285f4',  
    'Financials': '#ea4335',               
    'Industrials': '#fbbc04',           
    'Information Technology': '#34a853',  
    'Health Care': '#ff6d01',             
    'Consumer Staples': '#46bdc6',        
    'Real Estate': '#8e7dc3',             
    'Utilities': '#a64d79',               
    'Energy': '#ea95ea',                
    'Materials': '#ff4081',               
    'Telecommunication Services': '#00bcd4'  
}

risk_return_df['Risk_Return_Ratio'] = risk_return_df['Daily Return'] / risk_return_df['Daily Risk']

top_10 = risk_return_df.nlargest(10, 'Risk_Return_Ratio')
bottom_10 = risk_return_df.nsmallest(10, 'Risk_Return_Ratio')

plt.figure(figsize=(16, 12))
for sector in unique_sectors:
    sector_data = risk_return_df[risk_return_df['Sector'] == sector]
    plt.scatter(sector_data['Daily Risk'], 
               sector_data['Daily Return'], 
               c=sector_colors[sector], 
               label=sector, 
               alpha=0.7, 
               s=50)
for idx, row in top_10.iterrows():
    plt.annotate(row['Symbol'], 
                (row['Daily Risk'], row['Daily Return']),
                xytext=(5, 5), textcoords='offset points',
                fontsize=9, fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='lightgreen', alpha=0.7),
                ha='left')
for idx, row in bottom_10.iterrows():
    plt.annotate(row['Symbol'], 
                (row['Daily Risk'], row['Daily Return']),
                xytext=(5, -15), textcoords='offset points',
                fontsize=9, fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='lightcoral', alpha=0.7),
                ha='left')

aapl_data = risk_return_df[risk_return_df['Symbol'] == 'AAPL']
plt.scatter(aapl_data['Daily Risk'].iloc[0], 
           aapl_data['Daily Return'].iloc[0], 
           c='#34a853',  
           s=200,  
           marker='*',  
           edgecolors='black',
           linewidth=1,
           label='AAPL',
           zorder=10) 
plt.annotate('AAPL', 
            (aapl_data['Daily Risk'].iloc[0], aapl_data['Daily Return'].iloc[0]),
            xytext=(10, 10), textcoords='offset points',
            fontsize=9, fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.4', facecolor='yellow', alpha=0.9, edgecolor='black'),
            ha='left',
            zorder=11)

plt.xlabel('Risk', fontsize=16)
plt.ylabel('Return', fontsize=16)
plt.grid(True, alpha=0.3)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.tight_layout()
```
