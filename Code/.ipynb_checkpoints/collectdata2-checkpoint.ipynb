```python
#@title Packages

import pandas as pd
import numpy as np
import pandas_datareader.data as web
from datetime import datetime, timedelta
import re
import warnings

warnings.simplefilter(action='ignore', category=pd.errors.PerformanceWarning)
```

```python
#@title Read Datasets

stock_data = pd.read_csv("/Users/pedroalexleite/Desktop/Tese/Dados/dataset1.csv", decimal=',')

print(stock_data.head)
print(stock_data.shape)
print(stock_data.dtypes)
```

```python
#@title Remove 1259th Lines

#Dataset1
n = 1258
indices_to_drop = range(n - 1, len(stock_data), n)
stock_data = stock_data.drop(indices_to_drop).reset_index(drop=True)

print(stock_data.head)
print(stock_data.shape)
print(stock_data.dtypes)
```

```python
#@title Convert the Strings to Floats

stock_data['Date'] = pd.to_datetime(stock_data['Date'])

columns_to_convert = ["Open", "High", "Low", "Close", "Volume"]
for column in columns_to_convert:
    stock_data[column] = stock_data[column].str.replace(',', '.').astype(float)

print(stock_data.head)
print(stock_data.shape)
print(stock_data.dtypes)
```

```python
#@title Take out the Hour from Date

stock_data['Date'] = pd.to_datetime(stock_data['Date'], errors='coerce')
stock_data['Date'] = stock_data['Date'].dt.strftime('%Y/%m/%d')

print(stock_data.head)
```

```python
#@title Add Symbols

df_basic = pd.read_csv("/Users/pedroalexleite/Desktop/Tese/Dados/sp500.csv")
symbols = df_basic['Symbol'].unique()

repeated_symbols = [symbol for symbol in symbols for _ in range(1257)]
stock_data['Symbol'] = repeated_symbols

print(stock_data.head)
```

```python
#@title Drop Symbols Without Data

stock_data = stock_data.dropna(subset=['Open'])

print(stock_data["Symbol"].nunique())
```

```python
#@title Imput Weekends with the Values in the Last Week Day

def fill_missing_dates(group):
    group['Date'] = pd.to_datetime(group['Date'], errors='coerce') 
    if group['Date'].isna().all():  
        return group 
    full_date_range = pd.date_range(
        start=group['Date'].min(), 
        end=group['Date'].max(), 
        freq='D'
    )
    full_group = pd.DataFrame({'Date': full_date_range})
    full_group = pd.merge(full_group, group, on='Date', how='left')
    full_group = full_group.ffill()
    return full_group

stock_data = stock_data.groupby('Symbol', group_keys=False).apply(fill_missing_dates)
print(stock_data.head(10))
```

```python
import pandas as pd
import yfinance as yf

indicators = {
    "10Y Treasury Yield": "^TNX", 
    "Crude Oil Prices": "CL=F",     
    "Gold Prices": "GC=F",         
    "Consumer Confidence Proxy": "XLY",  
    "Retail Sales Proxy": "XRT",  
}

# Create an empty DataFrame with dates as index first
start_date = "2019-11-01"
end_date = "2024-11-01"
date_range = pd.date_range(start=start_date, end=end_date, freq='D')
macro_df = pd.DataFrame(index=date_range)

# Download and add each indicator
for name, ticker in indicators.items():
    data = yf.download(ticker, start=start_date, end=end_date, interval="1d")
    macro_df[name] = data['Close']

# Forward fill missing values
macro_df = macro_df.ffill()

# Merge with stock data
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
merged_data = pd.merge(stock_data, macro_df, how='left', left_on='Date', right_index=True)
stock_data = merged_data
```

```python
#@title Add Technical Indicators

def calculate_indicators(group):
    #trend
    group['SMA'] = ta.SMA(group['Close'], timeperiod=7)
    group['MACD'], group['MACD_Signal'], group['MACD_Hist'] = ta.MACD(group['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
    group['ADX'] = ta.ADX(group['High'], group['Low'], group['Close'], timeperiod=7)

    #volatility
    group['UPPER_BAND'], group['MIDDLE_BAND'], group['LOWER_BAND'] = ta.BBANDS(group['Close'], timeperiod=10, nbdevup=2, nbdevdn=2, matype=0)
    group['ATR'] = ta.ATR(group['High'], group['Low'], group['Close'], timeperiod=7)

    #momentum
    group['RSI'] = ta.RSI(group['Close'], timeperiod=7)
    group['STOCH_k'], group['STOCH_d'] = ta.STOCH(group['High'], group['Low'], group['Close'], fastk_period=7, slowk_period=3, slowd_period=3)
    group['CCI'] = ta.CCI(group['High'], group['Low'], group['Close'], timeperiod=10)
    group['ROC'] = ta.ROC(group['Close'], timeperiod=7)

    #volume
    group['OBV'] = ta.OBV(group['Close'], group['Volume'])
    group['AD'] = ta.AD(group['High'], group['Low'], group['Close'], group['Volume'])
    group['ADOSC'] = ta.ADOSC(group['High'], group['Low'], group['Close'], group['Volume'], fastperiod=3, slowperiod=10)

    #price
    group['SAR'] = ta.SAR(group['High'], group['Low'], acceleration=0.02, maximum=0.2)

    #statistical
    group['LINEARREG'] = ta.LINEARREG(group['Close'], timeperiod=7)

    #additional
    group['AROON_UP'], group['AROON_DOWN'] = ta.AROON(group['High'], group['Low'], timeperiod=7)
    group['AROONOSC'] = group['AROON_UP'] - group['AROON_DOWN']
    group['MFI'] = ta.MFI(group['High'], group['Low'], group['Close'], group['Volume'], timeperiod=7)

    return group

stock_data = stock_data.groupby('Symbol').apply(calculate_indicators)
stock_data = stock_data.reset_index(drop=True)
stock_data = stock_data.drop(columns=['MACD_Signal'])
stock_data = stock_data.drop(columns=['MACD_Hist'])
stock_data = stock_data.drop(columns=['UPPER_BAND'])
stock_data = stock_data.drop(columns=['LOWER_BAND'])
stock_data = stock_data.drop(columns=['STOCH_d'])
stock_data = stock_data.drop(columns=['AROON_UP'])
stock_data = stock_data.drop(columns=['AROON_DOWN'])
```
