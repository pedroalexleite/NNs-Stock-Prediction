```python
#@title Get the Average Sentimental Values of AV News for each Stock and Date

def calculate_average_sentiment(link):
    response = requests.get(link)
    data = response.json()

    if 'feed' in data:
        sentiment_scores = []
        for article in data['feed']:
            overall_sentiment_score = article.get('overall_sentiment_score')
            if overall_sentiment_score is not None:
                sentiment_scores.append(overall_sentiment_score)

        if sentiment_scores:
            avg_sentiment = sum(sentiment_scores) / len(sentiment_scores)
            return avg_sentiment
        else:
            return None
    else:
        return None
```

```python
#@title Fetch With Multitple Keys

def fetch_with_multiple_keys(links, delay=60, batch_size=100):
    av_sentiments = []
    for i in range(0, len(links), batch_size):
        batch = links[i:i + batch_size]

        for link in batch:
            api_key = random.choice(api_keys)
            link_with_key = link.replace('YOUR_API_KEY', api_key)
            average_sentiment = calculate_average_sentiment(link_with_key)
            av_sentiments.append(average_sentiment)

        print(f"Processed batch {i // batch_size + 1}/{len(links) // batch_size + 1}")
        time.sleep(delay)  

    return av_sentiments
```

```python
links = []
for index, row in stock_data.iterrows():
    symbol = row['Symbol']
    start_date = row['Date2'].strftime('%Y-%m-%d')
    end_date = row['Date'].strftime('%Y-%m-%d'))
        
    links.append(f'https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers={symbol}&from={start_date}&to={end_date}&apikey=YOUR_API_KEY')

av_sentiments = fetch_with_multiple_keys(links, delay=60, batch_size=100)
stock_data['AV Sentiment'] = av_sentiments
stock_data.to_csv('/Users/pedroalexleite/Desktop/Thesis/Data/AlphaVantageSentiment.csv', index=False)
```
