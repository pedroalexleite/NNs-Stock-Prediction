```python
#@title Packages

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.colors as mcolors
import seaborn as sns
from scipy.stats import zscore
from scipy.stats import mstats
from scipy.signal import savgol_filter
from scipy.signal import periodogram
from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm
from statsmodels.tsa.stattools import kpss
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import joblib
import os
import warnings
```

```python
#@title Basic Information

stock_data = pd.read_csv("/Users/pedroalexleite/Desktop/Tese/Dados/dataset3.csv")
```

```python
#@title Remove the First 2 Months

stock_data = stock_data[stock_data['Date'] >= '2020-06-01']

print(f"Number of Observations and Columns: {stock_data.shape}")
print(f"Number of stocks: {stock_data['Symbol'].nunique()}")
print("Variables:")
for column in stock_data.columns:
    print(column)
```

```python
#@title Pie Chart with the Sectors

sp500 = pd.read_csv("/Users/pedroalexleite/Desktop/Tese/Dados/sp500.csv")
stock_data_2 = stock_data.merge(sp500[['Symbol', 'Sector']], on='Symbol', how='left')
stock_data_unique = stock_data_2[['Symbol', 'Sector']].drop_duplicates(subset='Symbol').copy()
sector_counts = stock_data_unique['Sector'].value_counts()

strong_colors = ['#4285f4', '#ea4335', '#fbbc04', '#34a853', '#ff6d01', '#46bdc6', '#8e7dc3', '#a64d79', "#ea95ea", '#ff4081', '#00bcd4', '#795548']
plt.figure(figsize=(12, 8))
wedges, texts, autotexts = plt.pie(sector_counts, autopct='%1.1f%%', startangle=90, colors=strong_colors[:len(sector_counts)], pctdistance=0.75)
plt.legend(wedges, sector_counts.index, title="Sectors", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.axis('equal')
plt.show()
```

```python
#@title Pie Chart with the Variables per Category

categories = ['Univariate', 'Market', 'Technical']
values = [1, 5, 19]
strong_colors = ['#fbbc04', '#ea4335','#4285f4']

plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=strong_colors, pctdistance=0.85, labeldistance=1.1)
plt.axis('equal')
plt.show()
```

```python
#@title Missing Values

nan_counts = stock_data.apply(lambda x: x.isnull().sum())
nan_counts = nan_counts[nan_counts > 0]
print(nan_counts)
```

```python
#@title Correlation Between Variables

custom_cmap = mcolors.LinearSegmentedColormap.from_list(
    "custom_cmap", ['#4285f4', '#ea4335']
)

stock_data2 = stock_data.drop(columns=['Symbol', 'Date'])
corr_matrix = stock_data2.corr(numeric_only=True)

plt.figure(figsize=(12, 10))
sns.heatmap(
    corr_matrix,
    annot=False,
    cmap=custom_cmap,
    fmt='.2f',
    linewidths=0.5,
    cbar_kws={"orientation": "vertical"}
)
plt.xticks(rotation=90, fontsize=6)
plt.yticks(fontsize=8)
plt.gca().tick_params(axis='x', which='both', bottom=True, top=False)
plt.gca().spines['top'].set_visible(False)
labels = [item.get_text() for item in plt.gca().get_xticklabels()]
plt.gca().set_xticklabels(labels, fontsize=6)
plt.show()
```

```python
#@title Price through Time (Average Close Price for All Stocks)

stock_data['Date'] = pd.to_datetime(stock_data['Date'])
average_close = stock_data.groupby('Date')['Close'].mean()
plt.figure(figsize=(10, 6))
plt.plot(average_close, label='Average Close')
plt.xlabel('Date')
plt.ylabel('Average Close Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#@title Price through Time (AAPL Close Price)

aapl_data = stock_data[stock_data['Symbol'] == 'AAPL']
plt.figure(figsize=(10,6))
plt.plot(aapl_data['Date'], aapl_data['Close'], label='AAPL Close Price')
plt.xlabel('Date')
plt.ylabel('APPL Close Price')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
#@title Seasonal Decompose (Average Close Price for All Stocks)

stock_data['Date'] = pd.to_datetime(stock_data['Date'])
avg_close = stock_data.groupby('Date')['Close'].mean().sort_index()
result = seasonal_decompose(avg_close, model='additive', period=252)

fig = result.plot()
ax = fig.axes[-1]  

ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

ax.xaxis.set_minor_locator(mdates.MonthLocator())
ax.tick_params(axis='x', which='minor', bottom=False, labelbottom=False)
plt.setp(ax.get_xticklabels(), rotation=0, ha='center')
plt.show()
```

```python
#@title Seasonal Decompose (AAPL Close Price)

stock_data['Date'] = pd.to_datetime(stock_data['Date'])
aapl_data = stock_data[stock_data['Symbol'] == 'AAPL']
aapl_close = aapl_data.groupby('Date')['Close'].mean().sort_index()

result = seasonal_decompose(aapl_close, model='additive', period=252)

fig = result.plot()
ax = fig.axes[-1]

ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

ax.xaxis.set_minor_locator(mdates.MonthLocator())
ax.tick_params(axis='x', which='minor', bottom=False, labelbottom=False)
plt.setp(ax.get_xticklabels(), rotation=0, ha='center')
plt.show()
```

```python
#@title Weekly and Monthly Seasonality Heatmap (Average Close Price for All Stocks)

stock_data['Day of Week'] = stock_data['Date'].dt.dayofweek
stock_data['Month'] = stock_data['Date'].dt.month
weekly_avg = stock_data.groupby('Day of Week')['Close'].mean()
monthly_avg = stock_data.groupby('Month')['Close'].mean()

plt.figure(figsize=(10, 6))
sns.heatmap(weekly_avg.values.reshape(1, -1), cmap='coolwarm', xticklabels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(monthly_avg.values.reshape(1, -1), cmap='coolwarm', xticklabels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()

stock_data = stock_data.drop(columns=['Day of Week'])
stock_data = stock_data.drop(columns=['Month'])
```

```python
#@title Weekly and Monthly Seasonality Heatmap (AAPL Close Price)

aapl_data = stock_data[stock_data['Symbol'] == 'AAPL'].copy()
aapl_data['Day of Week'] = aapl_data['Date'].dt.dayofweek
aapl_data['Month'] = aapl_data['Date'].dt.month
weekly_avg = aapl_data.groupby('Day of Week')['Close'].mean()
monthly_avg = aapl_data.groupby('Month')['Close'].mean()

plt.figure(figsize=(10, 6))
sns.heatmap(weekly_avg.values.reshape(1, -1), cmap='coolwarm', xticklabels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(monthly_avg.values.reshape(1, -1), cmap='coolwarm', xticklabels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
#@title Periodogram for Seasonality Detection (Average Close Price for All Stocks)

stock_data['Date'] = pd.to_datetime(stock_data['Date'])
avg_close = stock_data.groupby('Date')['Close'].mean().sort_index()

frequencies, power_spectrum = periodogram(avg_close, scaling='density')
plt.figure(figsize=(10, 6))
plt.semilogy(frequencies, power_spectrum, color='blue')
plt.xlabel('Frequency (cycles per day)', fontsize=12)
plt.ylabel('Power Spectral Density', fontsize=12)
plt.grid(True)
plt.show()
```

```python
#@title Periodogram for Seasonality Detection (AAPL Close Price)

stock_data['Date'] = pd.to_datetime(stock_data['Date'])
aapl_data = stock_data[stock_data['Symbol'] == 'AAPL']
aapl_close = aapl_data.groupby('Date')['Close'].mean().sort_index()

frequencies, power_spectrum = periodogram(aapl_close, scaling='density')
plt.figure(figsize=(10, 6))
plt.semilogy(frequencies, power_spectrum, color='blue')
plt.xlabel('Frequency (cycles per day)', fontsize=12)
plt.ylabel('Power Spectral Density', fontsize=12)
plt.grid(True)
plt.show()
```

```python
#@title ACF and PACF (Average Close Price for All Stocks)

stock_data['Date'] = pd.to_datetime(stock_data['Date'])
avg_close = stock_data.groupby('Date')['Close'].mean().sort_index()

plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plot_acf(avg_close, ax=plt.gca(), lags=60)
plt.title('ACF')

plt.subplot(1, 2, 2)
plot_pacf(avg_close, ax=plt.gca(), lags=60)
plt.title('PACF')

plt.tight_layout()
plt.show()
```

```python
#@title ACF and PACF (AAPL Close Price)

aapl_data = stock_data[stock_data['Symbol'] == 'AAPL'].copy()
aapl_data['Date'] = pd.to_datetime(aapl_data['Date'])
aapl_close = aapl_data.groupby('Date')['Close'].mean().sort_index()

plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plot_acf(aapl_close, ax=plt.gca(), lags=60)
plt.title('ACF')

plt.subplot(1, 2, 2)
plot_pacf(aapl_close, ax=plt.gca(), lags=60)
plt.title('PACF')

plt.tight_layout()
plt.show()
```

```python
#@title Split Data into Train/Test (80/20)

min_date = stock_data["Date"].min()
max_date = stock_data["Date"].max()
split_date = min_date + pd.Timedelta(days=int(0.8 * (max_date - min_date).days))
train_data = stock_data[stock_data['Date'] < split_date]
test_data = stock_data[stock_data['Date'] >= split_date]
```

```python
#@title Save CSV Files

stock_data.to_csv('/Users/pedroalexleite/Desktop/Tese/Dados/dataset4.csv', index=False)
train_data.to_csv('/Users/pedroalexleite/Desktop/Tese/Dados/train.csv', index=False)
test_data.to_csv('/Users/pedroalexleite/Desktop/Tese/Dados/test.csv', index=False)
```
