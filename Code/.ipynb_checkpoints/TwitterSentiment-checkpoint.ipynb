```python
#@title Packages

import pandas as pd
import requests
from bs4 import BeautifulSoup
import yfinance as yf
from datetime import datetime, timedelta
import time
from textblob import TextBlob
import snscrape
import snscrape.modules.twitter as sntwitter
import certifi
import ssl
import urllib3


http = urllib3.PoolManager(cert_reqs='CERT_REQUIRED', ca_certs=certifi.where())
```

```python
#@title Dataframe with S&P500 Stocks and Their Basic Information

url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
tables = pd.read_html(url)
df_basic = tables[0]
df_basic = df_basic[['Symbol', 'Security', 'GICS Sector', 'GICS Sub-Industry', 'Headquarters Location']]

#print(df_basic['Symbol'].unique())
```

```python
#@title Today's Date and the Date From 1 Year Ago

today = datetime.today().date()
one_year_ago = today - timedelta(days=365)

print("Today's Date:", today)
print("Date From 1 Year Ago:", one_year_ago)
```

```python
#@title Dataframe with the Historical Data with 5 Days Intervals in a Space of 1 Year

stock_data = pd.DataFrame()

for symbol in df_basic['Symbol']:
    try:
        data = yf.download(symbol, start=one_year_ago, end=today, interval='5d')

        if data.empty or data.isnull().all().all():
            print(f"No data for {symbol}, skipping.")
            continue

        data['Symbol'] = symbol

        stock_data = pd.concat([stock_data, data])

    except Exception as e:
        print(f"Error downloading {symbol}: {e}")
```

```python
#@title Remove Unnecessary Variables

columns_to_remove = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
stock_data = stock_data.drop(columns=columns_to_remove)
stock_data.reset_index(inplace=True)
print("Head(5):")
print(stock_data.head(5))
print()
print("Shape:")
print(stock_data.shape)
print()
```

```python
#@title Scrape Tweets for a Specific Stock and Date

def scrape_tweets(symbol, date):
    date_str = date.strftime('%Y-%m-%d')
    query = f"{symbol} since:{date_str} until:{date_str} lang:en"
    tweets = []

    for tweet in sntwitter.TwitterSearchScraper(query).get_items():
        tweets.append({
            'symbol': symbol,
            'date': date_str,
            'tweet': tweet.content,
            'username': tweet.user.username,
            'tweet_id': tweet.id,
            'tweet_date': tweet.date,
            'retweets': tweet.retweetCount,
            'likes': tweet.likeCount,
            'replies': tweet.replyCount
        })

    return tweets
```

```python
#@title Scrape Tweets for a Specific Stock and Date in our Dataset

def scrape_all_tweets(stock_data):
    all_tweets = []

    for index, row in stock_data.iterrows():
        symbol = row['Symbol']
        date = row['Date']

        try:
            tweets = scrape_tweets(symbol, date)
            all_tweets.extend(tweets)

        except snscrape.base.ScraperException as e:
            print(f"Error scraping tweets for {symbol} on {date}: {e}")
            time.sleep(60)
            continue

        if index % 100 == 0 and index != 0:
            print(f"Scraped {index} out of {len(stock_data)} rows...")
            pd.DataFrame(all_tweets).to_csv('scraped_tweets_progress.csv', index=False)

        time.sleep(5)

    tweets_df = pd.DataFrame(all_tweets)

    return tweets_df

tweets_df = scrape_all_tweets(stock_data.head(10))
tweets_df.head(10)
```
