```python
#@title Packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
import joblib

np.random.seed(42)
tf.random.set_seed(42)
```

```python
#@title Read Data

#read files
train = pd.read_csv("/Users/pedroalexleite/Desktop/Tese/Dados/train.csv")
test =  pd.read_csv("/Users/pedroalexleite/Desktop/Tese/Dados/test.csv")
```

```python
#@title Prepare Data 

#we're going to use only 'AAPL'
train = train[train['Symbol'] == 'AAPL'].copy()
test = test[test['Symbol'] == 'AAPL'].copy()

def prepare_data(train, test, variable):
    #we're going to use only the price variable
    train = train[['Symbol', 'Date', variable, 'Regression Target']].copy()
    test = test[['Symbol', 'Date', variable, 'Regression Target']].copy()

    #normalize 
    numeric_cols = train.select_dtypes(include=['float64', 'int64']).columns
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaler.fit(train[numeric_cols])
    train[numeric_cols] = scaler.transform(train[numeric_cols])
    test[numeric_cols] = scaler.transform(test[numeric_cols])
```

```python
#@title Prepare Data for "Close"

prepare_data(train, test, 'Close')
```

```python
#@title Creating the Sequences

window_size = 7

def create_sequences(data, window_size):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size]) 
        y.append(data[i+window_size])  
        
    return np.array(X), np.array(y)

def prepare_stock_sequences(df, window_size):
    X_list, y_list = [], []
    for symbol, group in df.groupby('Symbol'):
        close_prices = group['Close'].values
        targets = group['Regression Target'].values
        
        if len(close_prices) < window_size + 1:
            continue
        
        X, y_seq = create_sequences(close_prices, window_size)
        y = targets[window_size:] 
        
        if len(X) == len(y):
            X_list.append(X)
            y_list.append(y)

    if not X_list or not y_list:
        raise ValueError("No valid sequences found. Check your data or window size.")
    
    X = np.vstack(X_list)
    y = np.concatenate(y_list)
    
    return X, y

X_train, y_train = prepare_stock_sequences(train, window_size)
X_test, y_test = prepare_stock_sequences(test, window_size)
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

```python
#@title Building

def my_model():
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
    model.add(Dropout(0.2)) 
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(units=1)) 
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
    model.summary()

my_model()
```

```python
#@title Training

history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

```python
#@title Predicting

predictions = model.predict(X_test)
```

```python
#@title De-Normalization

test_actual = test["Close"].values 
close_min = scaler.data_min_[numeric_cols.get_loc("Close")]
close_max = scaler.data_max_[numeric_cols.get_loc("Close")]
test_actual_de_normalized = test_actual * (close_max - close_min) + close_min
de_normalized_predictions = predictions.flatten() * (close_max - close_min) + close_min
```

```python
#@title Evaluating 

plt.figure(figsize=(10, 6))
plt.plot(test_actual_de_normalized, color='blue', label='Actual Price')
plt.plot(de_normalized_predictions, color='orange', label='Predicted Price')
plt.title('Actual vs Predicted Prices')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```
